**Data Spliting**

```python
x= st_data.iloc[:,:-1].values
y= st_data.iloc[:, 1].values

print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

**Importing Alogrithm library**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
lr=LinearRegression()
lr.fit(x_train , y_train)
y_predict=lr.predict(x_test)
print(y_predict)
```

```python
print(y_test)
```

```python
line=lr.coef_*x+lr.intercept_
plt.scatter(x,y)
plt.plot(x,line)
plt.show()
```

**Comparing actual vs predicted**

```python
print("Actual value",y_test)
print("Predicted value",y_predict)

```

```python
hours_st=9.5
hours_st=np.array(hours_st)
hours_st=hours_st.reshape(-1,1)
my_predict=lr.predict(hours_st)
print("No of hours studies:",format(hours_st))
print("Score he will after that hrs of study:",format(my_predict))
```

**Calulating the accuracy**

```python
mse=mean_squared_error(y_test,y_predict)
print(mse)
```
